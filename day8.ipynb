```python
import numpy as np
import pandas as pd
```

```python
with open("input/day8.txt", "r") as file:
    contents = file.read()
```

```python
moves, nodes = contents.split("\n\n")

```

```python
len(moves)
```

```python
nodes = nodes.split("\n")
```

```python
punkty = [x.replace("(",'').replace(')', '').split()[0] for x in nodes]
punkty
```

```python
kierunki = [x.replace("(",'').replace(')', '').replace(',','').split()[2:] for x in nodes]
kierunki
```

```python
dane = dict(zip(punkty, kierunki))
dane
```

```python
dane.get('AAA')
```

```python
current = 'AAA'
i = 0
ile_krokow = 0
while current != 'ZZZ':
    print('current', current)

    if moves[i] == 'R':
        current = dane.get(current)[1]
    elif moves[i] == 'L':
        current = dane.get(current)[0]
    i += 1
    ile_krokow += 1
    if i==len(moves):
        i=0

print('current ',current)
print('ile krokow ',ile_krokow)
```
