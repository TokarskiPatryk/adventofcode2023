```python
# mapa jest globalna
def is_digit_at_index(text, index):
    pattern = r'\d+'
    
    # Search for the pattern starting from the specified index
    matches = re.findall(pattern, text[:index + 1][::-1] + text[index:])
    
    if matches:
        return ''.join(matches[::-1])  # Reverse the order of matches to get the correct number
    else:
        return None
```
