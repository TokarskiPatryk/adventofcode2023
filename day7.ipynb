```python
def custom_sort(lst):
    order = {'A': 14, 'Q': 12, 'K': 13, 'J': 1, 'T': 10, '9': 9, '8': 8, '7': 7, '6': 6, '5': 5, '4': 4, '3': 3, '2': 2}
    #for hand in lst:
    #    print(hand[0])

    lista = sorted(lst, key=lambda x: [order[char] for char in x[0]])#, reverse=True)
    #print(lista)
    return lista

# Example usage:
cards = [('22345', 1), ('2J332', 1), ('AAAAA', 1), ('TTTT2', 1)]
print(cards)
sorted_cards = custom_sort(cards)
print(sorted_cards)
print(sorted_cards[0])
```
