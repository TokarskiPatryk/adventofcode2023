```python
df = pd.read_csv('input/day4.txt', 
    header=None, 
    sep=',', 
    index_col=0,
    names=['card no', 'win', 'my'])

win = df['win'].str.split(r'\s+', expand=True)
my = df['my'].str.split(r'\s+', expand=True)
```

```python
# 6 or 11 columns
row_win_values = win.iloc[0, 1:11].values
# first row from my
kolumn_win = len(row_win_values)+1
row_my_values = my.iloc[0, 1:].values
print(len(row_my_values))
print(row_my_values)
print(len(row_win_values))
print(row_win_values)
```
