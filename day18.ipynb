```python
import numpy as np
```

```python
with open("input/day18.txt") as f:
    df = [x.split(' ') for x in f.read().split('\n')]

df = np.array(df)
```

```python
warunki = df[:,0:2]
```

```python
zaznaczanie = np.zeros((1000,1000))
zaznaczanie[500,500] = 1
x1, y1 = (500,500)
for (x, y) in warunki:
    y = int(y)
    if x == 'R':
        zaznaczanie[x1, y1+1:y1+y+1] = 1

        y1+=y
    elif x == 'L':
        zaznaczanie[x1, y1-y:y1] = 1

        y1-=y
    elif x == 'U':
        zaznaczanie[x1-y:x1, y1] = 1

        x1-=y
    elif x == 'D':
        zaznaczanie[x1+1:x1+y+1,y1] = 1

        x1+=y
zaznaczanie[0:10,0:7]
```

```python
# find punkt najbardziej na prawo z 1 i w dol z 1
x2, y2 = np.where(zaznaczanie == 1)
x1 = x2.min()
y1 = y2.min()
x2 = x2.max()
y2 = y2.max()
with open("input/output.txt", 'w') as f:
    np.savetxt(f, zaznaczanie[x1-1:x2+1,y1-1:y2+1], fmt='%i')
```

```python
# save array zaznaczanie[x1:x2+1, y1:y2+1] as image without imshow
import matplotlib.pyplot as plt
plt.imshow(zaznaczanie[x1-1:x2+1, y1-1:y2+1])
plt.savefig('input/output.png')
```
