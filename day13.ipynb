```python
def row_check(mapa):
    for i in range(len(mapa)-1):
        # print(f'iteracja {i}')
        for j in range(0, len(mapa)-i-1):
            # print(f'j = {j}')
            if mapa[i-j] != mapa[j+i+1]:
                break
            # print(f'{i},{j}: {mapa[i-j]}  i {mapa[j+i+1]} równe')
            # sprawdzanie czy jest to ostatni element z gory i dolu
            
            if i-j==0 or j==len(mapa)-i-2:
                return [True, i+1]
    return [False, 0]
```

```python
file = [p.split() for p in open("input/day13.txt").read().split('\n\n')]

def num_of_diff(line1, line2, diff):
    d = 0
    for a, b in zip(line1, line2):
        if a != b: d += 1
        if d > diff: break
    return d

def find_mirror(patt, diff=0):
    for r in range(len(patt)-1):
        d = 0
        for i in range(min(r+1, len(patt)-r-1)):
            d += num_of_diff(patt[r-i], patt[r+1+i], diff)
            if d > diff: break
        else:
            if d == diff:
                return r+1, 0
    return find_mirror(list(zip(*patt)), diff)[::-1]

p1 = 0
p2 = 0
k = 0
for patt in file:
    k += 1
    r, c = find_mirror(patt, 0)
    p1 += c+100*r
    r, c = find_mirror(patt, 1)
    print(f'patt: {k} r: {r} c: {c}')
    p2 += c+100*r
print(p1, p2)
```
